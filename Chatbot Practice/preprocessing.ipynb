```python
import pandas as pd
import numpy as np
import chardet
```

```python
with open('NutritionalFacts_Fruit_Vegetables_Seafood.csv', 'rb') as f:
    encode = chardet.detect(f.read())
    
print("The encoding is: ",encode['encoding'])
```

```python
dataset = pd.read_csv("NutritionalFacts_Fruit_Vegetables_Seafood.csv", encoding='MacRoman')
dataset.head()
```

```python
dataset.info()
```

```python
df = dataset
```

```python
first_row = df.iloc[0].dropna()
for column in first_row.index:
    df.rename(columns={column: column + str(first_row[column])}, inplace=True)

df = df.drop(df.index[0])
df.head()
```

```python
# Converting to integer type
df.iloc[:, 2:23] = df.iloc[:, 2:23].astype(float)
df.info()
```
