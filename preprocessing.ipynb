```python
import pandas as pd
import numpy as np 
import chardet
```

```python
with open('NutritionalFacts_Fruit_Vegetables_Seafood.csv', 'rb') as f:
    encode = chardet.detect(f.read())
print("Encoding is :", encode['encoding'])
```

```python
dataset = pd.read_csv('NutritionalFacts_Fruit_Vegetables_Seafood.csv', encoding= 'MacRoman')
dataset.head()
```

```python
# Displaying the data type of each column
dataset.info()
```

```python
df = dataset
```

```python
first_row = df.iloc[0].dropna()
for column in first_row.index:
    df.rename(columns={column: column + str(first_row[column])}, inplace = True)
    
df = df.drop(df.index[0])
df.head()

# Combined the first row with the column name so it would be easier to display the data. It can also be useful for future uses
```

```python
# Now we convert some columsn to float type

df.iloc[:,3 : 23] = df.iloc[:,3 : 23].astype(float)
df.info()
```

```python
# Thats it for now, lets save it 
# not removing the null values as there are many items with null values and it is not due to mistake in data input
df.to_csv('clean_data.csv', index=False)
```
